```python
### 3. Economic Vulnerability by Geography

#This section examines how low-wage prevalence varies across regions.

#Higher values indicate a greater concentration of economically vulnerable individuals.
```

```python
import matplotlib.pyplot as plt

low_wage_geo = (
    df_clean.groupby('geo')['low_wage_flag']
    .mean()
    .sort_values(ascending=False)
)

plt.figure()
low_wage_geo.plot(kind='bar')
plt.title('Low-Wage Share by Geography')
plt.xlabel('Geography')
plt.ylabel('Proportion of Low-Wage Individuals')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
